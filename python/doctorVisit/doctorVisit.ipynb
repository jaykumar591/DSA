# import necessary linraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the data 

```python
data = pd.read_csv('./doctorVisit.csv')
```

# display the first few rows of the dataset

```python
print(data.head())
```

# Descriptive Statistics

```python
print('Descriptive statistics for the dataset:\n',data.describe())
```

# 2. Age and Health Relationship

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='health',y='age',data=data)

plt.title("Age and Health Status")
plt.xlabel('Health')
plt.ylabel('Age')
plt.show()
```

# Correlation between Age and health related columns

# 3. Income and health Relationship

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='income',y='health',hue='gender',data=data)

plt.title('Income vs Health Status by Gender')

plt.xlabel('Income')
plt.ylabel('Health')
plt.show()
```

# 4. Gender and Health Analysis

```python
plt.figure(figsize=(8,5))
sns.boxenplot(x='gender',y='health',data=data)
plt.title('Gender vs Health Status')
plt.xlabel('Gender')
plt.ylabel('Health')
plt.show()
```

# 5. Healthcare Utilization

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='private',y='visits',data=data)
plt.title('Healthcare Type (Private) vs Visits')
plt.xlabel('Private Healthcare')
plt.ylabel('Number of Visits')
plt.show()
```

# Free poor healthcare

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='freepoor',y='visits',data=data)
plt.title("Healthcare Type free poor vs Visits")
plt.xlabel('Free Poor Healthcare')
plt.ylabel('Number of  Visits')
plt.show()
```

# Free repatriate healthcare

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='freerepat',y='visits',data=data)
plt.title('Health Type Free Repatriate vs Visits')
plt.xlabel('Free repat Healthcare')
plt.ylabel('Number of Visits')
plt.show()
```

# 6 Chronic Illness Analysis

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['nchronic'], kde=True)
plt.title("Number of Chronic Illnesses")
plt.xlabel("Number of Chronic Illnesses")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['lchronic'], kde=True)
plt.title("Length of Chronic Illnesses")
plt.xlabel("Length of Chronic Illnesses")
plt.ylabel("Frequency")
plt.show()
```

# Relationship between chronic illnesses and reduced activity

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='reduced', y='nchronic', data=data)
plt.title("Reduced Activity vs Number of Chronic Illnesses")
plt.xlabel("Reduced Activity")
plt.ylabel("Number of Chronic Illnesses")
plt.show()
```

# 7. Income and healthcare Access

# relationship between income and healthcare access

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='private', y='income', data=data)
plt.title("Income and Private Healthcare Access")
plt.xlabel("Private Healthcare")
plt.ylabel("Income")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='freepoor', y='income', data=data)
plt.title("Income and Free Poor Healthcare Access")
plt.xlabel("Free Poor Healthcare")
plt.ylabel("Income")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='freerepat', y='income', data=data)
plt.title("Income and Free Repatriate Healthcare Access")
plt.xlabel("Free Repat Healthcare")
plt.ylabel("Income")
plt.show()
```

# 8. Health Stats and Illness

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='health', y='illness', data=data)
plt.title("Health Status vs Illness Count")
plt.xlabel("Health Status")
plt.ylabel("Illness Count")
plt.show()
```

```python
gender_counts = data['gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(7, 7))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Female vs Male Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
